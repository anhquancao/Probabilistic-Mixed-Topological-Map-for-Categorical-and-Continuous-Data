```python
def showImage(data):
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.imshow(data.reshape(16,15), cmap=cm.gray)
```

```python
def showIter(file, n):
    data = []
    with open(file, 'r') as csvfile:
            reader = csv.reader(csvfile, delimiter=',')
            for row in reader:
                data.append(list(map(lambda x: int(x), row)))
    data = np.array(data)

    fig, ax_arr = plt.subplots(n, n)
    for row in range(n):
        for col in range(n):
            ax_arr[row,col].imshow(data[row * n + col].reshape(16,15), cmap=cm.gray)
            ax_arr[row,col].axis("off")

```

```python
for i in range(0,21):
    file = '../out/exp Sat Sep 01 16:26:29 CEST 2018/binCentroid' + str(i)
    showIter(file, 5)
```
