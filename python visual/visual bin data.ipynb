```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import csv
import matplotlib.gridspec as gridspec
%matplotlib inline
```

```python
def showImage(data):
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.imshow(data.reshape(16,15), cmap=cm.gray_r)
```

```python
def showIter(file, n):
    data = []
    with open(file, 'r') as csvfile:
            reader = csv.reader(csvfile, delimiter=',')
            for row in reader:
                data.append(list(map(lambda x: int(x), row)))
    data = np.array(data)

    fig, ax_arr = plt.subplots(n, n)
    for row in range(n):
        for col in range(n):
            ax_arr[row,col].imshow(data[row * n + col].reshape(16,15), cmap=cm.gray_r)
            ax_arr[row,col].axis("off")
    return fig
```
