```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
probs = []
items=[]
labels=[]
iters=[]
for i in range(0,int(100/5)):
    i = i * 5 + 1
    iters.append(i)
    prob = pd.read_csv("./out/exp Wed Aug 01 16:34:05 ICT 2018/prob-" + str(i), header=None)
    probs.append(prob)
    item = pd.read_csv("./out/exp Wed Aug 01 16:34:05 ICT 2018/items-" + str(i), header=None)
    items.append(item)
    label = pd.read_csv("./out/exp Wed Aug 01 16:34:05 ICT 2018/labels-" + str(i), header=None,names=['x','y','label'])
    labels.append(label)
```

```python
for i in range(len(probs)):
    fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(10,3))
    
    sns.heatmap(probs[i], annot=True, ax=axes[0])
    axes[0].set_title("iter "+ str(iters[i]) + " p(c)")
    
    
    sns.heatmap(items[i], annot=True, ax=axes[1])
    axes[1].set_title("iter " +  str(iters[i]) + " Number of items assigned to each cell")
    
    sns.lmplot(x="x", y="y", data=labels[i], hue='label', fit_reg=False)
    plt.tight_layout()
    
```
