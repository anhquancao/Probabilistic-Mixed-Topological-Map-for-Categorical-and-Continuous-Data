```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
contData = pd.read_csv("./src/resources/d31/d31.csv", sep=',',names=['x', 'y'] )
```

```python
contData.head()
```

```python
dataSize = 3000
contData = contData.iloc[:dataSize,:]
```

```python
probs = []
items=[]
labels=[]
epsilons=[]
iters=[]
for i in range(0,int(30/2)):
    i = i * 2 + 1
    iters.append(i)
    prob = pd.read_csv("./out/exp Tue Aug 07 17:22:05 ICT 2018/prob-" + str(i), header=None)
    probs.append(prob)
    item = pd.read_csv("./out/exp Tue Aug 07 17:22:05 ICT 2018/items-" + str(i), header=None)
    items.append(item)
    label = pd.read_csv("./out/exp Tue Aug 07 17:22:05 ICT 2018/labels-" + str(i), header=None,names=['x','y','label'])
    labels.append(label)
    epsilon = pd.read_csv("./out/exp Tue Aug 07 17:22:05 ICT 2018/epsilon" + str(i), 
                          header=None).applymap(lambda x: round(x, 6))
    epsilons.append(epsilon)
```

```python
for i in range(len(probs)):
    fig, axes = plt.subplots(nrows=2, ncols=2,figsize=(10,8))
    
    sns.heatmap(probs[i], annot=False, ax=axes[0][0])
    axes[0][0].set_title("iter "+ str(iters[i]) + " p(c)")
    
    
    sns.heatmap(items[i], annot=False, ax=axes[0][1])
    axes[0][1].set_title("iter " +  str(iters[i]) + " Number of items assigned to each cell")
    
    sns.heatmap(epsilons[i], annot=False, ax=axes[1][0])
    axes[1][0].set_title("iter " +  str(iters[i]) + " Epsilon of Bernouli dist")
    
    sns.lmplot(x="x", y="y", data=labels[i], hue='label', fit_reg=False)
    
```
